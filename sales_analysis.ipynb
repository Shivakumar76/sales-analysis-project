```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')

# First look at the data
print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Check column names
print("Columns:\n", df.columns.tolist())

# Check data types & missing values
print("\nInfo:")
df.info()

# Basic statistics
print("\nBasic Stats:")
df.describe()
```

```python
# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Convert Order Date to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract year and month for later analysis
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

print("\nData cleaned successfully!")
df.head()
```

```python
# Total sales by category
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(category_sales)

# Bar chart
plt.figure(figsize=(8,5))
sns.barplot(x=category_sales.index, y=category_sales.values, palette='Blues_d')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.show()
```

```python
# Total sales by region
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print(region_sales)

# Bar chart
plt.figure(figsize=(8,5))
sns.barplot(x=region_sales.index, y=region_sales.values, palette='Greens_d')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.show()
```

```python
# Sales over time (monthly)
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
monthly_sales['Date'] = pd.to_datetime(monthly_sales[['Year', 'Month']].assign(Day=1))

# Line chart
plt.figure(figsize=(12,5))
plt.plot(monthly_sales['Date'], monthly_sales['Sales'], marker='o', color='steelblue')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Are high sales always profitable?
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category', alpha=0.6)
plt.title('Profit vs Sales by Category')
plt.xlabel('Sales ($)')
plt.ylabel('Profit ($)')
plt.tight_layout()
plt.show()
```

```python
# Top 10 best selling products
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print("Top 10 Products by Sales:")
print(top_products)
```

```python
# Most profitable sub-categories
sub_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=sub_profit.values, y=sub_profit.index, palette='RdYlGn')
plt.title('Profit by Sub-Category')
plt.xlabel('Total Profit ($)')
plt.ylabel('Sub-Category')
plt.tight_layout()
plt.show()
```

```python
# Loss-making sub-categories (negative profit)
print("\nLoss-making Sub-Categories:")
print(sub_profit[sub_profit < 0])
```

```python
# Loss-making sub-categories (negative profit)
print("\nLoss-making Sub-Categories:")
print(sub_profit[sub_profit < 0])
```

```python
print("=" * 50)
print("  SALES ANALYSIS SUMMARY")
print("=" * 50)

print(f"\n Total Orders: {df.shape[0]}")
print(f" Total Sales: ${df['Sales'].sum():,.2f}")
print(f" Total Profit: ${df['Profit'].sum():,.2f}")
print(f" Profit Margin: {(df['Profit'].sum()/df['Sales'].sum()*100):.2f}%")

print(f"\n Best Category: {category_sales.idxmax()} (${category_sales.max():,.2f})")
print(f" Best Region: {region_sales.idxmax()} (${region_sales.max():,.2f})")

print(f"\n Most Profitable Sub-Category: {sub_profit.idxmax()}")
print(f" Least Profitable Sub-Category: {sub_profit.idxmin()}")

print("\n" + "=" * 50)
print("  Analysis Complete!")
print("=" * 50)
```
